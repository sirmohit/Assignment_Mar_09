Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with
an example.

Ans:
    
    Probability Mass Function (PMF):
    The Probability Mass Function (PMF) is a function that gives the probability that a discrete random variable is exactly 
    equal to some value. In simpler terms, it provides the probability of each possible outcome of a discrete random variable.
    Mathematically, for a discrete random variable 
    X, the PMF is denoted by P(X=x).

    For a PMF:
        
    P(X=x)≥0 for all x (non-negativity).
    ∑P(X=x)=1 over all possible values of x (sums to 1).
    
    Probability Density Function (PDF):
    The Probability Density Function (PDF) is used for continuous random variables. Instead of giving the probability at
    a specific point (as in the PMF), the PDF provides the density of the probability at that point. In essence, the PDF
    represents a continuous probability distribution.

    For a PDF:
    
    f(x)≥0 for all x (non-negativity).
    ∫ −∞ to ∞ f(x)dx=1 (total area under the curve is 1).
    
    Example:
    PMF Example: Rolling a Fair Six-sided Die
    Let's say you roll a fair six-sided die. The possible outcomes are {1,2,3,4,5,6}.
    
    The PMF for this scenario is:P(X=1)=P(X=2)=P(X=3)=P(X=4)=P(X=5)=P(X=6)= 1/6
    
    Here, 
    P(X=x)=1/6 for each x between 1 and 6.
    
    PDF Example: Normal Distribution
    Consider a continuous random variable 
    X that follows a standard normal distribution (mean μ=0 and standard deviation σ=1).

    The PDF of the standard normal distribution is given by:
     
    f(x)=1/under root 2π * e^-x^2/2

    For a standard normal distribution, this equation defines the shape of the bell curve. If you were to calculate the 
    area under this curve between any two points a and b, you'd find the probability that X falls between a and b.





Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

Ans:
    
    Cumulative Density Function (CDF):
    The Cumulative Density Function (CDF) is a function that gives the probability that a random variable X takes on a value 
    less than or equal to x. In other words, the CDF provides the cumulative probability up to a certain value 
    x. For both discrete and continuous random variables, the CDF is defined as:
    
    Discrete Random Variable:
    F(x)=P(X≤x)=∑t≤x P(X=t)
    HereF(x) gives the cumulative probability that X is less than or equal to x.
    
    Continuous Random Variable:F(x)=P(X≤x)= ∫−∞ to ∞ f(t)dt
    Here, F(x) represents the area under the probability density function 

    f(t) from negative infinity up to 
    x.

    Example:
    CDF for a Fair Six-sided Die:
    Continuing with the example of rolling a fair six-sided die, the CDF F(x) at a particular point
    x would be:

    F(1)=P(X≤1)=1/6 
    F(2)=P(X≤2)=1/6 +1/6 =2/6
    F(3)=P(X≤3)=1/6+1/6+1/6=3/6 
    ...and so on until 
    F(6)=1.
    Here, 
    F(x) gives the cumulative probability of rolling a value less than or equal to x.
    
    Why CDF is used?
    Cumulative Information: The CDF provides a way to accumulate probabilities up to a particular value x, offering a
        comprehensive view of the distribution's behavior.

    Determine Percentiles: By examining the CDF, one can easily determine percentiles. For instance, the 50th percentile is
        the value x for which 0.5 F(x)=0.5, representing the median of the distribution.

    Comparison: CDFs are instrumental when comparing different distributions or when evaluating the performance of models
        in statistics or machine learning. Comparing CDFs can give insights into which distribution or model fits the data
        better.

    Computational Simplicity: In some cases, it's easier to work with the CDF rather than directly with the PDF or PMF, 
        especially for cumulative probabilities.


Q3: What are some examples of situations where the normal distribution might be used as a model?
Explain how the parameters of the normal distribution relate to the shape of the distribution.

    Examples of Situations for Using the Normal Distribution:
    The normal distribution, often referred to as the bell curve due to its characteristic symmetric shape, is widely used
    in various fields to model real-world phenomena. Here are some examples:

    Biological Measurements: Heights, weights, blood pressure, and other physiological measurements often follow a normal 
        distribution within specific populations.

    Educational Testing: Scores on standardized tests like the SAT or GRE are often assumed to follow a normal distribution.

    Quality Control: Measures of product dimensions or performance can often be modeled using the normal distribution,
        especially when the production processes are stable.

    Finance and Economics: Returns on assets in financial markets, such as stocks and bonds, often show a distribution that
        can be approximated by a normal distribution.

    Natural Phenomena: Many natural phenomena, such as measurement errors, tend to follow a normal distribution due to the
        central limit theorem, which states that the sum of a large number of independent random variables will tend to be
        normally distributed, regardless of their original distribution.

    Parameters of the Normal Distribution and Shape:
    The normal distribution is defined by two parameters: the mean μ and the standard deviation σ.

    Mean (μ):Represents the center or location of the distribution.
    The mean is the value around which the data points tend to cluster.
    Shifting the mean μ to the left or right will move the entire distribution horizontally along the x-axis.
    
    Standard Deviation (σ):

    Indicates the spread or dispersion of the data points around the mean.
    A smaller standard deviation means that the data points are closer to the mean, resulting in a narrower and taller
    bell curve.
    Conversely, a larger standard deviation indicates that the data points are more spread out from the mean, leading to
    a wider and flatter bell curve.

Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli
Distribution and Binomial Distribution?

Ans:
    
    Bernoulli Distribution:
    The Bernoulli distribution is a discrete probability distribution for a random variable that takes on only two possible
    outcomes, usually denoted as 1(success) and 0 (failure). The distribution is characterized by a single parameter 
    p, which represents the probability of success.

    For a Bernoulli random variable X:
    P(X=1)=p
    P(X=0)=1−p
    
    Example of Bernoulli Distribution:
    Consider a simple experiment of flipping a biased coin where the probability of getting a head (success) is p=0.7 and 
    the probability of getting a tail (failure) is 1−p=0.3.

    Here, a Bernoulli random variable can be defined as:
    X=1 if a head appears (with probability p=0.7)
    X=0 if a tail appears (with probability 1−p=0.3)
    
    Difference between Bernoulli Distribution and Binomial Distribution:
        
    Nature:

    Bernoulli Distribution: Represents a single trial with two possible outcomes (e.g., success or failure).
    Binomial Distribution: Represents the number of successes in n independent Bernoulli trials.
        
    Parameters:

    Bernoulli Distribution: Has a single parameter p, which is the probability of success in a single trial.
    Binomial Distribution: Has two parameters: n (number of trials) and p (probability of success in each trial).
            
    Random Variables:

    Bernoulli Distribution: Describes the outcome of a single Bernoulli trial (either success or failure).
    Binomial Distribution: Describes the total number of successes in n independent Bernoulli trials.

Q8: What is the z score? State the importance of the z score.

Ans:
    
    Z-Score:
    The Z-score, also known as the standard score, measures how many standard deviations a data point is from the mean of its
    distribution. Mathematically, the Z-score of an observation 

    X from a dataset with mean μ and standard deviation 
    σ is given by:Z= X−μ/σ

    Here:
    X is the individual data point.
    μ is the mean of the dataset.
    σ is the standard deviation of the dataset.
    
    Importance of the Z-Score:
    Standardization: One of the primary uses of the Z-score is to standardize or normalize data. By converting data into
        Z-scores, we transform the original data into a standard normal distribution with a mean of 0 and a standard
        deviation of 1. This standardization facilitates comparison between different datasets or observations that may
        have different units or scales.

    Identification of Outliers: Z-scores help in identifying outliers in a dataset. Observations with Z-scores far from
        zero (typically greater than 3 in absolute value) might be considered outliers because they lie significantly far
        from the mean and may represent unusual or anomalous observations.

    Probability Calculation: Z-scores are essential for computing probabilities associated with specific data values in 
        a normal distribution. By converting data into Z-scores, one can use standard normal (Z) tables or statistical 
        software to determine probabilities or percentiles associated with particular Z-values, facilitating hypothesis
        testing, confidence interval estimation, and other statistical analyses.

    Comparison and Standard Practice: In various fields like quality control, research, finance, and more, the Z-score
        serves as a universal measure to compare observations across different datasets or time periods. This
        standardization ensures consistent evaluation and interpretation of data, enabling informed decision-making.

Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

Ans:
    
    Central Limit Theorem (CLT):
    The Central Limit Theorem (CLT) is a fundamental theorem in probability theory and statistics that describes the shape of 
    the sampling distribution of the sample mean or sum. In simple terms, the CLT states that if you have a sufficiently large
    sample size n from any population (with a finite mean μ and a finite variance σ^2), the distribution of the sample mean 
    will be approximately normally distributed, regardless of the population distribution shape. Moreover, this approximation
    improves as the sample size n increases, even if the original population distribution is not normal.

    Mathematically, the CLT can be stated as follows:

    Let X1,X2,X3...Xn be a random sample of size n drawn from a population with mean μ and variance σ^2
    Then, as n approaches infinity:
        Xˉ=x1+x2+x3+...+xn/n
 
    has an approximately normal distribution with mean μ and standard deviation σ/under root n

    Significance of the Central Limit Theorem:
    Universal Applicability: The CLT is of paramount importance because it provides a foundation for using the normal 
    distribution as an approximation for the sampling distribution of the sample mean, irrespective of the shape of the 
    original population distribution. This universality allows statisticians and researchers to apply inferential 
    statistics confide
    

    Inference with Small Samples: Even with a small sample size, if the population is not normally distributed, the 
        sampling distribution of the sample mean tends toward a normal distribution as n increases, enabling valid
        statistical inference.

Q10: State the assumptions of the Central Limit Theorem.

Ans:
    
    The Central Limit Theorem (CLT) is a fundamental concept in statistics, but it comes with certain assumptions to hold 
    true. Here are the key assumptions of the Central Limit Theorem:

    Random Sampling: The samples are drawn randomly from the population. This means each member of the population has an
        equal chance of being selected, ensuring that the sample is representative of the population.

    Independence: Each observation within the sample is independent of the others. In practical terms, one observation's
        value does not influence or affect the value of another observation. This assumption ensures that the sample data
        points provide unique information.

    Finite Variance: The population from which the samples are drawn should have a finite variance. This assumption ensures
        that the variability within the population is bounded and does not grow indefinitely. In mathematical terms, both 
        the population mean and variance are finite.

    Sample Size: While the Central Limit Theorem is powerful and suggests that the sampling distribution of the sample mean 
        becomes approximately normal as the sample size n increases, the exact threshold for what constitutes a
        "sufficiently large" sample size depends on the population distribution. For many practical applications, a sample 
        size greater than 30 is often considered adequate, but this can vary based on the nature of the population 
        distribution.
